## Correlation woth Heatmap

A Correlation heatmap uses colored cell,typically in a monochromatic scale to show 2D correlation matrix(table) between two discreate dimensions or event ype. It is very important in Feature Selecetion

```python
sns.set_theme()
```

## Visualize the DATA

#### Relplot

```python
sns.relplot(data=df,x="total_bill",y="tip",col="time",hue="smoker",style="smoker",size="size")
```

```python
sns.relplot(data=dk,x="sepal_length",y="petal_length",hue="species",style="species")
```

#### Scatterplot

```python
sns.scatterplot(data=df,x="total_bill",y="tip",hue="smoker")
```

```python
sns.scatterplot(data=dk,x="sepal_length",y="petal_length",hue="species")
```

```python
sns.scatterplot(data=dk,x="sepal_length",y="petal_width",hue="species")
```
