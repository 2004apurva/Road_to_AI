### Bar Graph

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
language = ["English","Hindi","Chiniese","Urdu","Mongolien"]
People = [100,120,40,60,102]
ax.bar(language,People)
plt.xlabel("LANGUAGES")
plt.ylabel("PEOPLE")
plt.show()
```

### Pie Chart

```python
fig = plt.figure()
ay = fig.add_axes([0,0,1,1])
language = ["English","Hindi","Chiniese","Urdu","Mongolien"]
People = [100,120,40,60,102]
ay.pie(People,labels=language,autopct="%1.1f%%")
plt.show()
```

### Scatter Plot 

```python
t=np.linspace(0,10,30)
l = np.sin(t)
m = np.cos(t)
fig3 = plt.figure()
sc = fig3.add_axes([0,0,1,1])
sc.scatter(t,l,color="r")
sc.scatter(t,m,color="b")
plt.show()
```
